```python
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
from autoaim import Toolbox,ToolboxConfig,helpers
from toolz import pipe, curry
```

```python
plt.rcParams['figure.figsize'] = [10, 5]
```

```python
img = helpers.load('../../data/test16/110.jpeg')
```

```python
config = ToolboxConfig({'target_color': 'red','hsv_lower_value':90})
toolbox = Toolbox(config)
img1 = pipe(img,
     toolbox.split_channels,
     toolbox.preprocess,
     toolbox.binarize
)
img2 = pipe(img,
     toolbox.split_hsv
)
helpers.pltshow(img)
plt.show()
plt.imshow(img1)
plt.show()
plt.imshow(img2)
plt.show()
```
