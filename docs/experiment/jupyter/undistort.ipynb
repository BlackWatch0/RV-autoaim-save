```python
import sys
sys.path.append("../../")
from autoaim import helpers,Config,Toolbox
import cv2
import numpy as np
```

```python
img = helpers.load('../../data/test18/img10.jpg')
config = Config().data
toolbox = Toolbox(config)
camera_matrix = np.array(config['camera_matrix'])
distortion_coefficients = np.array(config['distortion_coefficients'])
```

```python
helpers.pltshow(img)
```

```python
img.shape[0:2]
```

```python
h, w = img.shape[:2]
new_camera_matrix, roi = cv2.getOptimalNewCameraMatrix(camera_matrix, distortion_coefficients, (w,h),1,(w,h))
dst = cv2.undistort(img, camera_matrix, distortion_coefficients, None, new_camera_matrix)
x, y, w, h = roi
dst = dst[y:y+h, x:x+w]
helpers.pltshow(dst)
```
